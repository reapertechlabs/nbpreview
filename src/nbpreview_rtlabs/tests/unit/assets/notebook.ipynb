```python
fmri = sns.load_dataset("fmri")
tips = sns.load_dataset("tips")
fmri.head()
```

```python
_, (ax_line, ax_box) = plt.subplots(ncols=2, facecolor="#1C1B1F", figsize=(20, 7))
(
    fmri.pipe(
        (sns.lineplot, "data"),
        x="timepoint",
        y="signal",
        hue="event",
        ax=ax_line,
        err_kws={"alpha": 1},
    ).set(
        xlabel=None,
        ylabel=None,
        yticks=[],
        xticks=[],
    )
)

(
    tips.pipe(
        (sns.boxplot, "data"),
        x="day",
        y="total_bill",
        hue="smoker",
        ax=ax_box,
    ).set(
        xlabel=None,
        ylabel=None,
        yticks=[],
        xticks=[],
    )
)
```
